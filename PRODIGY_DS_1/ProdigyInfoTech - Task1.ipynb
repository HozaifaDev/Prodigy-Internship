### Visualization

```python
df1.head()
```

```python
# Sum the values from '1960' to '2023' and create a new column 'Total Sum'
df1['Total Sum'] = df1.loc[:, 1960:2023].sum(axis=1)

# Get top 5 for 'Country Name'
top_countries = df1.groupby('Country Name')['Total Sum'].sum().nlargest(5)

# Get top 5 for 'Country Code'
top_codes = df1.groupby('Country Code')['Total Sum'].sum().nlargest(5)

# Get top 5 for 'Indicator Name'
top_indicators = df1.groupby('Indicator Name')['Total Sum'].sum().nlargest(5)
```

```python
import matplotlib.pyplot as plt

# Plot histograms
plt.figure(figsize=(14, 10))

# Overall histogram
plt.subplot(2, 2, 1)
plt.hist(df1['Total Sum'], bins=20, color='blue', alpha=0.7)
plt.title('Overall Distribution of Total Sum')

# Top 5 Country Names histogram
plt.subplot(2, 2, 2)
top_countries.plot(kind='bar', color='green', alpha=0.7)
plt.title('Top 5 Country Names by Total Sum')
plt.xticks(rotation=45)

# Top 5 Country Codes histogram
plt.subplot(2, 2, 3)
top_codes.plot(kind='bar', color='red', alpha=0.7)
plt.title('Top 5 Country Codes by Total Sum')
plt.xticks(rotation=45)

# Top 5 Indicator Names histogram
plt.subplot(2, 2, 4)
top_indicators.plot(kind='bar', color='orange', alpha=0.7)
plt.title('Top 5 Indicator Names by Total Sum')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
